# オブジェクトの可視化(matplot)

```python
import numpy as np
import matplotlib.pyplot as plt
def visualize_mesh(pos, face):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.axes.xaxis.set_ticklabels([])
    ax.axes.yaxis.set_ticklabels([])
    ax.axes.zaxis.set_ticklabels([])
    ax.plot_trisurf(
        pos[:, 0], pos[:, 1], pos[:, 2],
        triangles=face,
        antialiased=False,
        alpha=1,
        )
    plt.show()
```

# オブジェクトの可視化(plotly)

```python
import plotly.graph_objects as go
```

```python
def visualize_mesh_plotly(vertices=vertices, faces=faces, c=None):

    # エッジを取得
    xe, ye, ze = [], [], []
    for f in faces.numpy():
        v1 = vertices.numpy() [f[0]]
        v2 = vertices.numpy() [f[1]]
        v3 = vertices.numpy() [f[2]]
        xe.extend([v1[0], v2[0], v3[0], v1[0], None])
        ye.extend([v1[1], v2[1], v3[1], v1[1], None])
        ze.extend([v1[2], v2[2], v3[2], v1[2], None])

    # 描画
    fig = go.Figure()

    ## surface
    fig.add_trace(go.Mesh3d(
            x=vertices[:,0], y=vertices[:,1], z=vertices[:,2],
            i=faces[:,0], j=faces[:,1], k=faces[:,2],
            opacity=0.5,
            color='skyblue',
            contour = dict(color = 'red', show = True, width = 16),
    ))

    ## edge
    fig.add_trace(go.Scatter3d(
            x=xe, y=ye, z=ze,
            mode='markers+lines',
            line=dict(color="blue"),
            marker=dict(color="blue", size=2)
    ))

    fig.show()
```

```python
i=15
vertices = dataset.get_data(i).vertices
faces = dataset.get_data(i).faces
visualize_mesh_plotly(vertices, faces)
```

```python
import torch_geometric
print('pyg_ver:', torch_geometric.__version__)

from torch_geometric.data import Data, InMemoryDataset
```

```python
# 自作データセット用のクラス
# https://pytorch-geometric.readthedocs.io/en/latest/tutorial/create_dataset.html
# 上記URLで構築されているOwnDatasetクラスを参考
class MyDataset(InMemoryDataset):
    def __init__(self, root, transform=None, pre_transform=None, pre_filter=None):
        super().__init__(root, transform, pre_transform, pre_filter)
        self.data, self.slices = torch.load(self.processed_paths[0])

    @property
    def raw_file_names(self):
        return []

    @property
    def processed_file_names(self):
        return ['/tmp/mdata.pt']

    def download(self):
        pass

    def process(self):
        # Read data into huge `Data` list.
        # data_list = [...]

        data_list= []
        for i, d in enumerate(dataset_airplane):
            vertices, faces = d.data.vertices, d.data.faces
            data = Data(pos=vertices, face=faces.T, y=0, edge_attr=None)
            data_list.append(data)
            if i%10==0: print(f'index:{i} was loaded in OriginalDataset')
            if i==100: break

        for i, d in enumerate(dataset_car):
            vertices, faces = d.data.vertices, d.data.faces
            data = Data(pos=vertices, face=faces.T, y=1, edge_attr=None)
            data_list.append(data)

            if i%10==0: print(f'index:{i} was loaded in OriginalDataset')
            if i==100: break

        if self.pre_filter is not None:
            data_list = [data for data in data_list if self.pre_filter(data)]

        if self.pre_transform is not None:
            data_list = [self.pre_transform(data) for data in data_list]

        data, slices = self.collate(data_list)

        torch.save((data, slices), self.processed_paths[0])

# transformを定義
import torch_geometric.transforms as T
transform = T.Compose([
    #T.face_to_edge.FaceToEdge(remove_faces=True),
    T.SamplePoints(2**12),
    T.KNNGraph(6),
])

new_dataset = MyDataset(root='./data/orig/', transform=transform)

new_dataset[0]
```
